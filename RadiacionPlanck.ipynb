## Radiación de cuerpo Negro

### Ley de Radiación de Planck

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import FormatStrFormatter

from math import pi
h=6.62607015e-34 # constante de Planck
c=299792458      # velocidad de la luz
k=1.3806488e-23  # constante de Boltzmann
T=1250           # Temperatura
A=2*np.pi*h*c**2
L=np.arange(200e-9,6e-6,10e-9)
L5=L**5
hc=h*c
LkT=L*k*T
B=hc/LkT
I=A/(L5*(np.exp(B)-1))
fig, ax = plt.subplots()
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.plot(L,I,'r-')
plt.show()
```
